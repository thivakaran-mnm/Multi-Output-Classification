```python
hash_field_train = train_data[hash_columns].to_dict(orient='records')
hash_field_test = test_data[hash_columns].to_dict(orient='records')
```

```python
# Use FeatureHasher to create numerical features from hash field
from sklearn.feature_extraction import FeatureHasher

hasher = FeatureHasher()
```

```python
# Transform hash fields into numerical features

hashed_features_train = hasher.transform(hash_field_train)
hashed_features_test = hasher.transform(hash_field_test)
```

```python
# Convert hashed features to DataFrames
hashed_features_train_df = pd.DataFrame.sparse.from_spmatrix(hashed_features_train)
hashed_features_test_df = pd.DataFrame.sparse.from_spmatrix(hashed_features_test)
```

```python
# Drop the original hash field from train and test datasets
train_features = train_data.drop(columns=hash_columns)
test_features = test_data.drop(columns=hash_columns)
```

```python
# Combine hashed features with original features
train_features = pd.concat([train_features, hashed_features_train_df], axis=1)
test_features = pd.concat([test_features, hashed_features_test_df], axis=1)
```

```python
# Identify feature types
numerical_features = train_features.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = train_features.select_dtypes(include=['object']).columns.tolist()
boolean_features = train_features.select_dtypes(include=['bool']).columns.tolist()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Define preprocessing for numerical, categorical, and boolean features
numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
boolean_transformer = OneHotEncoder(drop='if_binary')
```

```python
bow_pipeline = Pipeline(steps=[
    ('bow', CountVectorizer()),  # BoW vectorization
])
```

```python
tfidf_pipeline = Pipeline(steps=[
    ('tfidf', TfidfVectorizer()),  # TF-IDF vectorization
])
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from gensim.models import Word2Vec

# Custom transformer for Word2Vec embeddings
class Word2VecTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, vector_size=100, window=5, min_count=1, sg=1):
        self.vector_size = vector_size
        self.window = window
        self.min_count = min_count
        self.sg = sg
        self.model = None

    def fit(self, X, y=None):
        sentences = [sentence.split() for sentence in X]
        self.model = Word2Vec(sentences, vector_size=self.vector_size, window=self.window, min_count=self.min_count, sg=self.sg)
        return self

    def transform(self, X):
        def get_vector(sentence):
            words = sentence.split()
            vector = np.mean([self.model.wv[word] for word in words if word in self.model.wv], axis=0)
            if isinstance(vector, np.ndarray):
                return vector
            else:
                return np.zeros(self.vector_size)

        return np.array([get_vector(sentence) for sentence in X])
```

```python
word2vec_pipeline = Pipeline(steps=[
    ('word2vec', Word2VecTransformer())
])
```

```python
# Combine Bag of Words and TF-IDF pipelines using FeatureUnion
text_transformer = ColumnTransformer(
    transformers=[
        ('bow', bow_pipeline, 'text_column'),
        ('tfidf', tfidf_pipeline, 'text_column'),
        ('word2vec', word2vec_pipeline)
    ])
```

```python
# Create a column transformer that applies the appropriate transformations
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features),
        ('bool', boolean_transformer, boolean_features)
    ])
```

```python
# Create a pipeline that includes preprocessing and the classifier
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', MultiOutputClassifier(RandomForestClassifier()))
])
```

```python
num_samples = train_features.shape[0]
train_labels_reduced = train_labels.iloc[:num_samples]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X_train, X_val, y_train, y_val = train_test_split(train_features, train_labels_reduced, test_size=0.2, random_state=42)

# Align train_labels with train_features by index
y_train_r = y_train.loc[X_train.index]
y_val_r = y_val.loc[X_val.index]

# Fit the pipeline on the training data
pipeline.fit(X_train, y_train_r)

# Predict on validation set
y_val_pred = pipeline.predict(X_val)

# Generate the classification report
print(classification_report(y_val_r, y_val_pred, target_names=[f'y{i+1}' for i in range(y_val_pred.shape[1])]))
```
